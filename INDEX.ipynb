# INDEX — FDA Products (Drug Names)

שליפת עמודת **DrugName** מקובץ `data/Drugs/Products.txt` (FDA Orange Book).

```python
from pathlib import Path

import pandas as pd

PRODUCTS_PATH = Path("data/Drugs/Products.txt")

products = pd.read_csv(PRODUCTS_PATH, sep="\t", dtype=str)
drug_names = products["DrugName"]
unique_drug_names = drug_names.drop_duplicates().reset_index(drop=True)

print(f"Total rows: {len(drug_names):,}")
print(f"Unique drug names: {len(unique_drug_names):,}")
unique_drug_names
```

```python
OUTPUT_PATH = Path("data/Drugs/unique_drug_names.csv")

unique_drug_names.to_frame().to_csv(OUTPUT_PATH, index=False)
print(f"Saved {len(unique_drug_names):,} rows to {OUTPUT_PATH.resolve()}")
```
