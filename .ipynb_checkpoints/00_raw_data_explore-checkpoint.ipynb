````python

from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 200)

RAW_DIR = Path("data/raw")

files = sorted(RAW_DIR.iterdir())
print(f"Folder: {RAW_DIR.resolve()}")
print(f"Number of files: {len(files)}\n")

for i, f in enumerate(files, 1):
    size_mb = f.stat().st_size / (1024 * 1024)
    print(f"{i:2}. {f.name}  ({size_mb:.1f} MB)")
````

## 2. ניתוחים_260415.xlsx

````python
# load
surgeries = pd.read_excel(RAW_DIR / "ניתוחים_260415.xlsx")
surgeries.head(1)
````

````python
# shape and column names
print("shape:", surgeries.shape)
print("\ncolumns:")
for col in surgeries.columns:
    print(col)
````

````python
surgeries.head()
````

````python
eda_summary = pd.DataFrame({
    "column": surgeries.columns,
    "dtype": surgeries.dtypes.astype(str),
    "missing_count": surgeries.isnull().sum(),
    "missing_pct": (surgeries.isnull().mean() * 100).round(2),
    "n_unique": surgeries.nunique(dropna=True)
})

display(eda_summary.sort_values("missing_pct", ascending=False))
````

````python
# age
print("min:", surgeries["גיל"].min())
print("max:", surgeries["גיל"].max())
print("mean:", round(surgeries["גיל"].mean(), 2))
````

````python
# gender (% of rows)
gender_pct = surgeries["מין"].value_counts(normalize=True) * 100
print(gender_pct.round(2))
````

````python
# date range
start = surgeries["תאריך ניתוח"].min()
end = surgeries["תאריך ניתוח"].max()
years = (end - start).days / 365.25

print("first surgery:", start)
print("last surgery:  ", end)
print("total span (years):", round(years, 2))
````

````python
# cases per parent department (א/ב → parent name)
dept = surgeries["מחלקה מנתחת"].str.strip()
parent_dept = dept.str.replace(r" [אב]$", "", regex=True)

dept_cases = surgeries.groupby(parent_dept)["מספר מקרה"].nunique()
total_cases = surgeries["מספר מקרה"].nunique()

dept_summary = pd.DataFrame({
    "cases": dept_cases,
    "pct": (dept_cases / total_cases * 100).round(2),
})

dept_summary.sort_values("cases", ascending=False)
````

### Charts

````python
# date range over time — unique cases per month, by parent department
monthly = (
    surgeries.assign(parent_dept=parent_dept, month=surgeries["תאריך תחילת ניתוח"].dt.to_period("M"))
    .groupby(["month", "parent_dept"])["מספר מקרה"]
    .nunique()
    .unstack(fill_value=0)
)

monthly.plot(figsize=(12, 5), marker=".")
plt.title("Unique cases per month by department")
plt.xlabel("Month")
plt.ylabel("Unique cases")
plt.legend(title="Department")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# unique cases per parent department
dept_summary["cases"].sort_values().plot(kind="barh", figsize=(8, 4), color="steelblue")
plt.title("Unique cases per department")
plt.xlabel("Unique מספר מקרה")
plt.ylabel("Department")
plt.tight_layout()
plt.show()
````

### Unique cases & cross-department overlap

````python
# how many unique cases?
n_rows = len(surgeries)
n_cases = surgeries["מספר מקרה"].nunique()

print("total rows:", n_rows)
print("unique מספר מקרה:", n_cases)
print("avg rows per case:", round(n_rows / n_cases, 2))

# can one case appear in more than one department?
dept = surgeries["מחלקה מנתחת"].str.strip()
parent_dept = dept.str.replace(r" [אב]$", "", regex=True)

depts_per_case = (
    surgeries.assign(parent_dept=parent_dept)
    .groupby("מספר מקרה")["parent_dept"]
    .nunique()
)

multi_dept_cases = depts_per_case[depts_per_case > 1]
print("\ncases in more than one parent department:", len(multi_dept_cases))

if len(multi_dept_cases) > 0:
    dept_combos = (
        surgeries.assign(parent_dept=parent_dept)
        .groupby("מספר מקרה")["parent_dept"]
        .apply(lambda x: " + ".join(sorted(x.unique())))
    )
    print("\nwhere (parent department combinations):")
    print(dept_combos[depts_per_case > 1].value_counts())
````

## 3. סוג דם גובה משקל BMI.xlsx

````python
# load
bmi = pd.read_excel(RAW_DIR / "סוג דם גובה משקל BMI.xlsx")
bmi.head(1)
````

````python
# shape and column names
print("shape:", bmi.shape)
print("\ncolumns:")
for col in bmi.columns:
    print(col)
````

````python
bmi.head()
````

````python
bmi[['גובה','משקל','BMI']].describe()
````

````python
bmi.shape
````

````python
eda_summary_bmi = pd.DataFrame({
    "column": bmi.columns,
    "dtype": bmi.dtypes.astype(str),
    "missing_count": bmi.isnull().sum(),
    "missing_pct": (bmi.isnull().mean() * 100).round(2),
    "n_unique": bmi.nunique(dropna=True),
})

display(eda_summary_bmi.sort_values("missing_pct", ascending=False))
````

````python
# outliers — height, weight, BMI
print(bmi[["גובה", "משקל", "BMI"]].describe())

height_bad = (bmi["גובה"] == 0) | (bmi["גובה"] < 100) | (bmi["גובה"] > 220)
weight_bad = (bmi["משקל"] < 20) | (bmi["משקל"] > 300)
bmi_bad = (bmi["BMI"] < 10) | (bmi["BMI"] > 80)

print("\nheight == 0:", (bmi["גובה"] == 0).sum())
print("height < 100 or > 220:", height_bad.sum())
print("weight < 20 or > 300 kg:", weight_bad.sum())
print("BMI < 10 or > 80:", bmi_bad.sum())

print("\nexamples — extreme weight:")
bmi.nlargest(5, "משקל")[["מספר מקרה", "גובה", "משקל", "BMI"]]
````

````python
# join coverage vs surgeries
surgery_cases = set(surgeries["מספר מקרה"].dropna())
surgery_patients = set(surgeries["patient"].dropna())

bmi_cases = set(bmi["מספר מקרה"].dropna())
bmi_patients = set(bmi["Patient"].dropna())

print("מספר מקרה:")
print("  in BMI:       ", len(bmi_cases))
print("  in surgeries: ", len(surgery_cases))
print("  overlap:      ", len(bmi_cases & surgery_cases))
print("  surgeries only:", len(surgery_cases - bmi_cases))
print("  BMI only:     ", len(bmi_cases - surgery_cases))

print("\nPatient:")
print("  in BMI:       ", len(bmi_patients))
print("  in surgeries: ", len(surgery_patients))
print("  overlap:      ", len(bmi_patients & surgery_patients))
print("  surgeries only:", len(surgery_patients - bmi_patients))
print("  BMI only:     ", len(bmi_patients - surgery_patients))

# duplicate rows per case
rows_per_case = bmi.groupby("מספר מקרה").size()
print("\ncases with more than 1 row:", (rows_per_case > 1).sum())
````
